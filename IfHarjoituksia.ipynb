```python
num1 = input("give number 1")
num2 = input("give number 2")
num3 = input("give number 3")

try:
    val1 = int(num1)
    val2 = int(num2)
    val3 = int(num3)

    if val1 == val2 or val2 == val3:
        print("numbers 1 and 2 or numbers 2 and 3 are equal")
    else:
        print("numbers are not equal")

except ValueError:
    print("not an int")
    
```
